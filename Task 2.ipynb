# Student Result Management System

This program helps teachers:
- Add student records
- View all students
- Update scores
- Search students
- Calculate grades
- Display class statistics

```python
#Students record stored in a dictionary

students = {"Oloko Sultan": 60, "Olasupo Toheeb": 98, "Akajiobi Emmanuel": 80, "Abdulwarith Warith": 60}
```

```python
# Add student

def add_student():
    name = input("Enter student name: ")

    try:
        score = float(input("Enter student score: "))

        if score < 0 or score > 100:
            print("Score must be between 0 and 100.")
            return

        students[name] = score
        print("Student added successfully.")

    except ValueError:
        print("Invalid score entered.")
```

```python
add_student()
```

```python
students
```

```python
# Function to calculate grade

def calculate_grade(score):
    if score >= 70:
        return "A"
        
    elif score >= 60:
        return "B"
        
    elif score >= 50:
        return "C"
        
    elif score >= 45:
        return "D"
        
    else:
        return "F"
```

```python
#view_students

def view_students():

    print("STUDENTS RECORD")
    
    for name, score in students.items():

        grade = calculate_grade(score)

        print(f"Name: {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")
        print("-" * 40)
    
    if not students:
        print("No student records found.")
        return
```

```python
view_students()
```

```python
# update_score

def update_score():

    name = input("Enter student name: ")

    if name in students:

        try:
            score = float(input("Enter new score: "))

            if score < 0 or score > 100:
                print("Score must be between 0 and 100.")
                return

            students[name] = score
            print("Score updated successfully!")

        except ValueError:
            print("Invalid score entered.")

    else:
        print("Student not found.")
```

```python
update_score()
```

```python
#search_student

def search_student():

    name = input("Enter student name: ")

    if name in students:

        score = students[name]

        grade = calculate_grade(score)

        print("Student Found")
        
        print("-" * 30)
        print(f"Name: {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")

    else:
        print("Student not found.")
```

```python
search_student()
```

```python
#class_statistics

def class_statistics():

    if not students:
        print("No student records available.")
        return

    scores = list(students.values())

    print("CLASS STATISTICS")
    print("Highest Score:", max(scores))
    print("Lowest Score:", min(scores))
    print("Average Score:", round(sum(scores) / len(scores), 2))
    print("Total Students:", len(scores))
```

```python
class_statistics()
```

```python
while True:

    print("\n===== STUDENT RESULT MANAGEMENT SYSTEM =====")
    print("1. Add Student")
    print("2. View Students")
    print("3. Update Score")
    print("4. Search Student")
    print("5. Class Statistics")
    print("6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        view_students()

    elif choice == "3":
        update_score()

    elif choice == "4":
        search_student()

    elif choice == "5":
        class_statistics()

    elif choice == "6":
        print("Program Closed.")
        break

    else:
        print("Invalid choice. Please try again.")
```

# INVENTORY MANAGEMENT SYSTEM

This program helps staff members to:

- Store products and quantities.
- Add new products.
- Update product quantities.
- Search for products.
- Remove products.
- Display inventory records.
- Generate inventory statistics.
- Identify low-stock items.


```python
#INVENTORY RECORDS STORED IN A DICTIONARY

products ={"milk":20, "golden more":59, "corn flakes":67, "garri":70}

products
```

```python
#ADD PROducts

def add_product():
    
    product_name = input("Enter a products: ")
    
    try:
        quantity = int(input("Enter the available quantity: "))

        if quantity < 0:
            print("Quantity cannot be negative")
            return

        products[product_name] = quantity

        print("product added successfully")
        
    except ValueError:
        print("Enter a valid number")
```

```python
add_product()
```

```python
products
```

```python
#VIEW PRODUCTS

def view_inventory():
    
    print("INVENTORY RECORDS")
    
    for product_name, quantity in products.items():

        print(f"There is {quantity} packs of {product_name}")

    if not products:
        print("There is no product available")
        return
```

```python
view_inventory()
```

```python
#UPDATE PRODUCT QUANTITTY

def update_product():
    product_name = input("Enter the product to be updated: ")

    if product_name in products:
        
        try:
            quantity = int(input("Enter the quantity of product available: "))

            if quantity < 0:
                print("Quantity cannot be negative")
                return
        
            products[product_name] = quantity

            print("product updated successfully")
            
        except ValueError:
            print("Invalid quantity entered")
        
    else:
        print("products not available in the stock ")

   
```

```python
update_product()
```

```python
#SEARCH PRODUCTS

def search_product():
    product_name = input("Enter a product: ")

    
    if product_name in products:

        quantity = products[product_name]
        
        print(f"There is {product_name} in stock")
        print(f"There is {quantity} cartons of {product_name} available")

    else:
        print("Product not available in the stock")
```

```python
search_product()
```

```python
#REMOVE PRODUCT

def remove_product():
    product_name = input("Enter product name")

    if product_name in products:
        del products[product_name]

        print(f"{product_name} removed from the stock succesfully")

    else:
        print(f"{product_name} is not in the stock")
```

```python
remove_product()
```

```python
#LOW STOCK REPORT

def low_stock_report():

    print("LOW STOCK REPORT")
    
    found = False
    
    
    for product_name, quantity in products.items():

        if quantity < 5:
            print(f"{product_name} : {quantity}")
            found = True

    if not found:
        print("No low stock found")        
```

```python
low_stock_report()
```

```python
#INVENTORY STATISTICS

def inventory_statistics():
    if not products:
        print("No inventory records available")
        return

    quantities = list(products.values())

    print("INVENTORY STATISTICS")
    print("Total number of products:", len(products.keys()))
    print("Highest products quantity:", max(products.values()))
    print("Lowest products quantity:", min(products.values()))
    print("Total stock available:", sum(products.values()))
```

```python
inventory_statistics()
```

```python
while True:

    print("\n===== INVENTORY MANAGEMENT SYSTEM =====")
    print("1. Add product")
    print("2. View inventory")
    print("3. Update product quantity")
    print("4. Search product")
    print("5. Remove products")
    print("6. Low stock report")
    print("7. Inventory statistics")
    print("8. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_product()

    elif choice == "2":
        view_inventory()

    elif choice == "3":
        update_product()

    elif choice == "4":
        search_product()

    elif choice == "5":
        remove_product()

    elif choice == "6":
        low_stock_report()

    elif choice == "7":
        inventory_statistics()

    elif choice == "8":
        print("Program Closed.")
        break

    else:
        print("Invalid choice. Please try again.")
```
